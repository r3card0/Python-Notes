En nuestra vida cotidiana seguimos caminos, procesos, rutas, lo que sea. Estos caminos en ocasiones abren otros caminos y entonces tenemos que preguntarnos por donde ir. Que camino elegimos? El A? o El B?, o tal vez haya un camino C. 

Normalmente, decidimos en base al beneficio que vayamos a obtener o para evitar algn suplicio. Nos decimos, si tomo este camino, debo tener esto, o si tomo este otro, a donde me podr llevar?

En programacin podemos hacer lo mismo, usando los comandos: if, elif, else.

Imagina que quieres salir de viaje y tienes las siguientes opciones: la ciudad de Berln en  Alemania, La ciudad de Nueva York en USA o el parque de la esquina. El lugar que visitars estar en funcin del dinero que logres ahorrar para pagar el destino deseado y disfrutar tu estancia. A continuacin, est la lista de precios de los destinos:

New York, USA = $400 USD

Berlin, Ale = $800 USD

Parque = menos de $400 USD

Fcilmente podemos decir que:
1. Si tu ahorro es igual o mayor a $800 USD, entonces tu destino es Berlin, pero 
2. si tu ahorro es igual o mayor a $400 USD y menor a $800 USD, entonces tu destino puede ser New York, USA, pero 
3. si no ahorras ms de $400 USD, entonces tendrs la opcin que queda. El confiable y nico parque de la esquina.

El siguiente razonamiento se puede expresar en la siguiente sintaxis: 

```python
# viaje 
ahorro = int(input('Cuanto dinero ahorraste durante el ao?: '))
berlin = 800
NewYork = 400
if ahorro >= berlin:
    print('Felicidades, los alemanes te esperan con los tarros llenos de cerveza. Iras a Berlin Alemania!')
elif NewYork < ahorro:
    print('La esttua de la libertad te saluda porque a Nueva York es a donde irs!')
else:
    print('Bueno, pues prepara los sandwiches y lleva repelente para mosquitos, porque el parque de la esquina, te espera!')
```

### El clsico, es mas grande que, menor que

En programacin es clasico hacer un programa que compare nmeros y te diga que valor es mas grande. En este ejemplo jugaremos con las condicionales if, elif y else agregando los booleanos: True y False

Si el valor de X es mayor que el valor de Y, entonces es verdadero, si no, es falso

supongamos que X es un valor fijo. X es igual a 2500. Y el valor de Y es un valor que eligiras

```python
Y = int(input('El valor de Y es mayor que X?. Elige el valor de -Y-: '))
X = 2500
if Y > X:
    print(True)
else:
    print(False)
```

De esta forma, la computadora nos responde dependiendo del valor que le asignemos.

Como podemos ver, la sintaxis es muy sencilla:

1. comando if
2. expresin, operacin
3. Usar ":" despues de la expresin
4. Indentacin -> 4 espacios de sangra.
2. resultado esperado si se cumple la condicion

### Conversor de monedas

Ahora tendremos un programa donde nos haga la conversin de monedas colombianas a dlares

El valor de pesos colombianos a dolares es igual a: 1 USD = 4015.1 COP(Colombian pesos)

fuente: [themoneyconverter](https://themoneyconverter.com/ES/USD/COP)

Puedes checar la fuente cada vez y actualizar el valor actual en el programa

```python
# conversor de monedas
cop = int(input('Cuantos pesos colombianos tienes?: '))
cop = float(cop)
dollar_price = 4015.1
dollars = cop / dollar_price
dollars = round(dollars, 2)
print('Tienes $' + str(dollars) + ' dolares')
```

Explicacin del programa por lnea:
1. *Elige una cantidad*. La variable **cop**, recibe la cantidad elegida a travs del comando *input* y el valor recibido se convierte a nmero entero con el comando *int*
2. *Convertir a decimal*. Una vez recibido el valor en la variable **cop**, la convierte en numero decimal con el comando *float*
3. *Precio actual del dolar*. El precio actual del dolar est asignado en la variable **dollar_price**
4. *Operacin de la conversin*. La variable **dollars** guarda la operacin entre los valores de las variables **cop** y **dollar_price**
5. *Redondeo del resultado*. Se redondea el resultado con el comando *round* y el nmero 2, indica el nmero de decimales.
6. *Se imprime el resultado*. Mediante el comando *print* se muestra el mensaje en la pantalla usando una concatenacin de datos de tipo numerico y de texto. Se hace la conversin a texto de la variable **dollars** con el comando *str*
