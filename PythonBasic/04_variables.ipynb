## Variables
Las variables son cajas o contenedores que se crean para guardar valores. Estos valores pueden contener los siguientes tipos de datos: textos, numricos, booleanos, etc.

Estos tipos de datos pueden ser usados para realizar diversas funciones en un programa.

```python
# 1 variable caja; valor 'Books'
caja = 'Books'
print(caja,'of Harry Potter')
```

```python
# 2
x = 20
y = 'Hola a todos'
print(x)
print(y)
```

Una variable es creada al momento de asignarle un valor. En el ejemplo # 1 la variable **caja** se le asign el valor **'Books'**. En el ejemplo # 2, la variable "**x**" se le asign el valor de **20** y la variable "**y**" se le asign el valor **'Hola a todos'**

El comando **print()** imprime los valores de las variables. Solo hay que poner el nombre de la variable entre parentesis.

### Formas de asignar valores

Python permite asignar valores a mltiples variables en una sola lnea.

```python
var1, var2, var3 = 'Tampico','Len','Mrida'
print(var1)
print(var2)
print(var3)
```

**Nota**: Asegrate que el nmero de variables corresponda con el nmero de valores, si no (else), se tendr un error

```python
var1, var2, var3 = 'Tampico', 'Len'
print(var1)
print(var2)
print(var3)
```

Tambin, Python permite guardar el mismo valor en distintas variables en una sola lnea

```python
var4 = var5 = var6 = 'Molcajete'
print(var4)
print(var5)
print(var6)
```

### Desempacar una coleccin de datos. Unpacking
Cuando se tiene una coleccin de datos, Python permite extraer los datos en variables. Es conocido como "unpacking", del ingls que significa "desempacar".

```python
frutas = ["manzana", "pia", "durazno"]
fruta1, fruta2, fruta3 = frutas
print(fruta1)
print(fruta2)
print(fruta3)
```

*Las colecciones de datos son listas de cosas*. En este caso es una lista comn de frutas. En Python las listas de cosas se colocan entre corchetes [ ]. Para la lista de frutas, se usaron las comillas dobles "" para identificar cada elemento/fruta y porque se trata de texto. El texto es uno de los tipos de datos que se maneja en Python. Python conoce el texto como *string*

### Variables de Salida /Output Variables

Como se ha visto, el comando **print()** es usado para imprimir los valores de las variables

Se puede combinar texto y el valor de una variable usando el smbolo "**+**"

```python
var1 = 'Tampico'
print(var1 + ' es una ciudad del estado de Tamaulipas')
```

Tambin se puede usar "+" para juntar valores de variables

```python
var1 = 'Tampico'
var2 = ' Hermoso'
print(var1 + var2)
```

Para el tipo de dato como los nmeros, el signo "**+**" se puede usar como operador aritmtico. 

```python
print(5 + 4)
print(100 + 10)
print(12345 + 768)
```

Si se quiere combinar tipos de dato de texto con nmeros, Python arrojar un error

```python
var1 = 'Tampico'
var2 = 5
print(var1 + var2)
```

Esto es porque el texto es un tipo de dato diferente a los nmeros. De manera directa, en Python no es permitido juntar manzanas con peras. Es decir, no combina texto con nmeros en un comando **print()** de forma directa.

Para combinar el texto con los nmeros, se debe hacer una conversin de los numeros. Esta conversion se hace con el comando "str". str significa string. 

```python
var1 = 'Tampico'
var2 = 5
print(var1 + str(var2))
```

Otra forma es realizar el cambio dentro de la misma variable var2 que contiene el valor 5

```python
var1 = 'Tampico'
var2 = 5
var2 = str(var2)
print(var1 + var2)
```

De esta forma, al usar el comando "str", se hace la conversion de numero a texto, convirtiendo la pera (numero 5) en manzana (texto). Y Python permitir la impresin del mensaje

Para que se vea ms esttico y dejar espacios entre los valores de las variables var1 y var2 se puede hacer lo siguiente:

Directamente dejar el espacio en var1

```python
var1 = 'Tampico '
var2 = 5
var2 = str(var2)
print(var1 + var2)
```

O tambin, dejar espacio dento del comando **print** usando el smbolo "**+**"

```python
var1 = 'Tampico'
var2 = 5
var2 = str(var2)
print(var1 + " " + var2)
```
