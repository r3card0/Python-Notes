## Comando **return**

En general, una funcin **toma** argumentos, **ejecuta** operaciones y **retorna** un resultado. El resultado que una funcin retorna es generalmente conocido como *valor de retorno de una funcin*. 

El comando **return** es un comando que se puede usar dentro de una funcin y se usa para devolver o retornar el resultado de la funcin. 

Sintaxis: palabra clave **return** seguida del valor de retorno.

    > return value



```python
# ejemplo de uso del comando return
def suma(a,b):
    resultado = a + b
    return resultado

suma(2,2)
```

```python
# retornando varios resultados
def operaciones(a,b):
    suma = a + b
    multiplicacion = a * b
    division = a / b
    return suma, multiplicacion, division

operaciones(1500, 350)
```

```python
# ejemplo 3 devolviendo un valor tipo string
def string(c):
    texto = c
    return texto

string('Hola Mundo')
```

```python
# ejemplo 4 devolviendo valores booleanos
def booleanos():
   x = 1500
   y = 1850
   if x >= y:
       return True
   else:
       return False

booleanos()
```

En los ejemplos 1 y 2 se observa que el comando **return** devuelve los valores de las variables asignadas, las cuales contienen las operaciones.

En el ejemplo 3,  el comando **return** retorna el string contenido en la variable.

En el ejemplo 4, el comando **return** se usa para devolver el valor despus de que una condicin sea validada.

## Entry Point

En programacin, un entry point o punto de entrada, es donde las primeras instrucciones de un programa sern ejecutadas y donde el programa tiene acceso a los argumentos de las lneas de comandos.

Es decir, todo programa requiere un entry point para ejecutarse.

El entry point usado en funciones es el siguiente:

    > if __name__=='__main__':

```python
# entry point
def saludo():
    print('Hola Mundo')

if __name__=='__main__':    # entry point
    saludo()                # lnea que invoca a la funcin saludo
```

Como se puede ver, en la linea siguiente del entry point, se requiere una indentacin y colocar la lnea que invoca la funcin 

Veamos el ejemplo del conversor de monedas y agregaremos el entry point en el lugar correspondiente.

```python
# agregar entry point
def opcion_elegida(pesos_locales,dollar_price):
    moneda_local = int(input('Cuantos ' + pesos_locales  + ' tienes?: '))
    moneda_local = float(moneda_local)
    dollars = round(moneda_local/dollar_price,2)
    print('Tienes $' + str(dollars) + ' dolares')
    

def elegir_opcion():
    menu = int(input(""" 
    Elige la moneda que deseas convertir
    Pesos colombianos   1
    Pesos Mexicanos     2
    Pesos Argentinos    3

    Elige una opcin: """))

    if menu == 1:
        opcion_elegida('pesos colombianos',4015.1)
    
    elif menu == 2:
        opcion_elegida('pesos mexicanos', 21.5)

    elif menu == 3:
        opcion_elegida('pesos argentinos', 102)
    
    else:    
        print('Elige una opcin correcta')


if __name__=='__main__':        # entry point
    elegir_opcion()
```

En programas sencillos, tal vez no se note la importancia de usar el entry point, sin embargo, en proyectos complejos donde se usen mltiples mdulos son muy tiles para asegurar el buen desempeo de los programas que se ejecutan.

Es por eso que hay que identificarlos y comenzar a usarlos.

- Las funciones ayudan para mejorar el control y mantemiento de un programa, reutilizar cdigo, evitar colapsos en los sistemas, facilitar la lectura de un cdigo y permite resolver problemas de forma rpida y efectiva
- Los parmetros son variables definidas en una funcin.
- Los parmetros guardan valores. A los valores se les llama *argumentos*.
- Los argumentos son los valores asignados a los parmetros de una funcin para que realicen tareas.
- El comando **return** es usado dentro de la funcin para retornar los valores de la funcin.
- El entry point es la primera instruccin que un programa usa para ejecutarse. Toda funcin debe tener un entry point para asegurar un buen rendimiento en proyectos complejos.
