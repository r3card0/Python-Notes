# Tipados en Python

## ¿Qué son los "tipados"?

* Los tipos son los *tipos de datos* que maneja un lenguaje de programación, como es Python. Los tipos son: **Strings** (textos), **booleanos**, **numéricos** (integros(int) y flotantes(float)) y los **arreglos** (listas, diccionarios, tuplas)
* Cada lenguaje de programación, maneja de forma particular a cada tipo.
* De esta forma, los lenguajes de programación, se pueden clasificar en base a como tratan a los tipos.
* Cada lenguaje, tendrá un **tipado** diferente
* Hay 4 tipos de tipado: estático, dinámico, débil y fuerte.

* Recordar el proceso que tiene un lenguaje compilado para funcionar.
* Se escriben las lineas de codigo posteriormente hay un **tiempo de compilación** donde el las lineas de codigos se pasan a lenguaje máquina el cual hace que se ejecuten en la computadora.
