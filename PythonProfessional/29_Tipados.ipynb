# Tipados en Python

## ¿Qué son los "tipados"?

* Los tipos son los *tipos de datos* que maneja un lenguaje de programación, como es Python. Los tipos son: **Strings** (textos), **booleanos**, **numéricos** (integros(int) y flotantes(float)) y los **arreglos** (listas, diccionarios, tuplas). Ver [tipos de datos](https://github.com/r3card0/Python-Notes/blob/main/PythonBasic/05_tipos-de-datos.ipynb)
* Cada lenguaje de programación, maneja de forma particular a cada tipo.
* De esta forma, los lenguajes de programación, se pueden clasificar en base a como tratan a los tipos.
* Cada lenguaje, tendrá un **tipado** propio del lenguaje
* Hay 4 tipos de tipado: estático, dinámico, débil y fuerte.


* Recordar el proceso que tiene un lenguaje compilado para funcionar.
* Se escriben las lineas de codigo posteriormente hay un **tiempo de compilación** donde el las lineas de codigos se pasan a lenguaje máquina el cual hace que se ejecuten en la computadora.



## Tipados

* Estáticos
* Dinámicos
* Débil
* Fuerte

### Estáticos

* Los lenguajes de tipado **estático** son los que muestran los **errores de tipos** en tiempo de compilación




* Esto se da cuando en el codigo, por error se trata una cadena de caracteres como número, el lenguaje indicara que hay un error en **tiempo de compilacion**, esto quiere decir que el programa no se ejecutará hasta que se resuelva el error de tipo.

* En Java cuando se pretende usar un numero en una variable tipo string, inmediatamente marca error

```bash
String str = "Hello"
str = 5
```

* En Python, cuando se asigna un valor numerico a una variable, este lo toma como numerico.

````python
var = 5
print(var)
````

````python
type(var)
````

* En Python cuando se usan las comillas simple o dobles, lo toma como una variable de tipo string

````python
var2 = '5'
print(var2)
````

````python
type(var2)
````

* Por lo tanto, se puede decir que **Python es un lenguaje de tipado Dinámico**

### Dinámicos

* Los lenguajes de **tipado dinámico** son los que muestran los **errores de tipo** en el **tiempo de ejecución**
* Python muestra loes errores de tipo en tiempo de ejecución

### Débil

* Los lenguajes de tipado *débil*, tratan a los tipos de una forma menos estricta

### Fuerte

* Los lenguajes de tipado **fuerte**, tratan a los tipos de forma muy estricta
* Python trata de manera severa a los datos de diferentes tipos
* Python **siempre marca error** ``` TypeError```cuando se pretende sumar variables de tipo numerico con variables de tipo string, porque esta operación no cumple con la lógica del programa.

````python
#por ejemplo
x = 5 # variable numérica
y = "7" # variable de texto
suma = x + y
````

* En cambio, lenguaje como **JavaScript** si permite ejecutar la logica, determinando concatenar o juntar los valores de forma automática. Esto hace de *JavaScript* un lenguaje de tipo débil
* En cambio en Python, para que se ejecute la logica, es necesario **castear**  uno de las variables.
* Es decir, cambiar el tipo de dato numérico ```x = 5 ```a tipo de dato de texto.
* El término *castear*, es más conocido como **casting** o **parser** en la comunidad de Python. [Ver tipos de datos](https://github.com/r3card0/Python-Notes/blob/main/PythonBasic/05_tipos-de-datos.ipynb)

````python
#por ejemplo
x = 5 # variable numérica
y = "7" # variable de texto
suma = str(x) + y  # aqui se cambia de numérico a texto
print(suma)
````

En conclusión, se puede decir que:
* **Python** es un lenguaje **fuerte** y **dinámico**
* Python es **fuerte** porque es muy severo en operaciones de diferentes tipos
* Python es **dinamico** porque marca errores en **tiempo de ejecución**

Es importante decir que ser un lenguaje de tipo dinámico es peligroso
