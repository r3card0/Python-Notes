# ¿Cómo funciona Python?
### Python es un lenguaje intepretado

* Python, normalmente es clasificado como un *lenguaje interpretado*, el cual compila el código fuente a un grupo de instrucciones para una máquina virtual y el interprete de Python es una implementación  de esa máquina virtual.

* Un lenguaje compilado se comunica directamente con la computadora. Es una comunicación directa con la computadora. La comunicación se lleva a cabo a través de la compilación.

### Lenguajes Intepretados
* Son leguajes en el su código fuente es traducido en instrucciones nativas del CPU a medida que se ejecuta el programa.

### Bytecode

* Es un lenguaje de bajo nivel, que puede ser leído por un interprete. El interprete de Python es una de las partes de este lenguaje de bajo nivel: Bytecode. No llega a ser código maquina, ósea cero y unos.

* Es leído por una máquina virtual. Esta habilidad, le permite que Python pueda funcionar en varios sistemas operativos, convirtiéndose así en un lenguaje multiplataforma. Ya que solo se requiere del lenguaje con su intérprete para ser leído en cualquier sistema operativo.

### Los lenguajes interpretados son más lentos? 
* Son más lentos que los lenguajes compilados. Por milisegundos.

### ¿Qué es el garbage collector?
* Lo que hace es tomar los objetos y las variables que no están en uso y las elimina. Python cuenta con una sección que es un garbage collector. 

### ¿Qué es la carpeta ````__pycache__````?
* ¿Cuándo se crea la carpeta en el proyecto, lo que hay en ella es el Bytecode. Tiene el intérprete. Además que mantiene en memoria el código que se ha escrito, permitiendo así tener más velocidad al momento de programar.

* El interprete de Python es leído por una máquina virtual y a través de esta, Python puede ser leído en los tres principales sistemas operativos: Windos, MacOS y Linux. Esto lo convierte en un lenguaje multiplataforma.
