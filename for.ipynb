Para usar el comando "**for**" primero se requiere una variable con un valor inicial.
1. Tener una variable con un valor inicial. Que sea integer
2. Tener una condición que frene el loop. por ejemplo, mientras i sea menor que 10, entonces que se siga ejecutando el ciclo
3. Saber como va ir creciendo



````python
for i in range(0,11):
  if i < 9:
    print(i)
````

Imprime línea por línea los valores del rango

````python
for i in range(0,5):              # para la variable i en el rango de 0 a 5
    print(i)                      # imprime los valores de i línea por línea
````

Imprime línea por línea los caracterés del texto 'Samantha'

````python
for i in 'Samantha':              # para i en 'Samantha'
  print(i)                        # imprime los valores de i línea por línea
````

Imprime línea por línea los caracterés del texto 'Samantha' y convertirlos a mayúsculas.

````python
for i in 'Samantha':              # para i en 'Samantha'
  print((i).upper())
````

````python
def run():
  for i in 'samantha':
    print((i).capitalize())


run()
````

Ejecuta el ciclo hasta que encuentres el nombre de Luis Eduardo

````python
list_kids = ['Aarón','Samanta','Saúl','César','Luis Eduardo','Mauricio','Daniel']
for kid in list_kids:         # para la varible kid en la lista kids
  print(kid)                  # imprime kid
  if kid == 'Luis Eduardo':   # si kid es igual a 'Luis Eduardo'
    break                     # entonces, rompe el ciclo
````

Print the 10 first square number, one per line. Start at 0, so the first square number is 02, followed by 12, 22, and so on up to 92.

````python
for i in range(0,10):   # para la variable i en el rango de 0 a 10
  print(i**2)           # imprime la potencia de 2, línea por línea
````
